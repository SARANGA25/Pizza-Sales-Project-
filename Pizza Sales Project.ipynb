### Add a | Hours | Minutes | Seconds | Columns

```python
df['order_time'] = df['order_time'].astype('string')
df[["hours", "minutes", "seconds"]] = df["order_time"].str.split(":", expand=True)
```

### Add a Day Columns

```python
df["day"] = pd.to_datetime(df["order_date"], errors='coerce')
df["day"] = df["day"].dt.day_name()
```

### Add a Month Columns 

```python
df['Month'] = df['order_date'].str[3:5].astype(int) 
```

### Month Trend For Total Orders

```python
Month_Trend = df.groupby('Month', as_index=False)['order_id'].nunique().sort_values(by='order_id', ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(data=Month_Trend, x='Month', y= 'order_id')
for bars in ax.containers:
    ax.bar_label(bars)
```

### Daily Trend For Total Orders

```python
Day_Trend = df.groupby('day',as_index=False)['order_id'].nunique().sort_values(by='order_id',ascending=False)
sns.set(rc={'figure.figsize':(12,5)})
ax=sns.barplot(data=Day_Trend, x='day', y='order_id')
for bars in ax.containers:
    ax.bar_label(bars)
```

### Percentage of Sales by Pizza Category

```python
sales_by_Category = df.groupby('pizza_category')['total_price'].sum()
#calculate Percentge
percentage_sales = (sales_by_Category/sales_by_Category.sum())*100
#Plot dount chart
plt.pie(percentage_sales, labels=percentage_sales.index, autopct='%1.1f%%', startangle=90, wedgeprops =dict(width=0.4))
plt.axis('equal')
plt.title('Percentage of Sales by Pizza Category',fontweight="bold")
plt.show()
```
